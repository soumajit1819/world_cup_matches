```python
import pandas as pd

# Load the dataset
file_path = '/content/WorldCupMatches (2).csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
```

```python
# Number of rows and columns
rows, columns = df.shape
rows, columns
```

```python
# Print specific range using slicing (first 10 rows as an example)
df.iloc[:10]
```

```python
# Summarization statistics
summary_stats = df.describe()
summary_stats
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Replace missing numerical values with mean/median
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    if df[column].isnull().sum() > 0:
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
#heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Convert date/time columns to datetime objects
# Replace 'Year' with the actual name of your date column
df['Year'] = pd.to_datetime(df['Year'], errors='coerce')

# Extract numerical features for correlation
numerical_df = df.select_dtypes(include=['number'])

corr = numerical_df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Example Barplot (Number of matches per year)
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Example Histogram (Distribution of goals scored)
plt.figure(figsize=(10, 6))
sns.histplot(df['Home Team Goals'], bins=20, kde=True)
plt.show()
```

```python
# Distribution Plot (Home Team Goals)
plt.figure(figsize=(10, 6))
sns.distplot(df['Home Team Goals'], kde=True)
plt.show()
```

```python
# Scatterplot (Home Team Goals vs Away Team Goals)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=df)
plt.show()
```
